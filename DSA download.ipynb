```python
import requests
import os

def download_pdf(url, folder_path):
    try:
        response = requests.get(url)
        response.raise_for_status()
    except requests.exceptions.HTTPError as errh:
        print ("Http Error:",errh)
        return False
    except requests.exceptions.ConnectionError as errc:
        print ("Error Connecting:",errc)
        return False
    except requests.exceptions.Timeout as errt:
        print ("Timeout Error:",errt)
        return False
    except requests.exceptions.RequestException as err:
        print ("OOps: Something Else",err)
        return False

    # Get the file name from the URL
    file_name = os.path.join(folder_path, url.split("/")[-1])

    with open(file_name, "wb") as pdf_file:
        pdf_file.write(response.content)
    return True

def download_pdfs_from_links(pdf_links, folder_path):
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)

    for idx, link in enumerate(pdf_links):
        print(f"Downloading PDF {idx + 1}/{len(pdf_links)}...")
        success = download_pdf(link, folder_path)
        if success:
            print("Downloaded successfully!")
        else:
            print("Failed to download the PDF.")

if __name__ == "__main__":
    # Provide your list of PDF links here
    pdf_links = [
        
"https://mahades.maharashtra.gov.in/files/publication/dsa/2021-22/DSA_2022_Thane.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa/2021-22/DSA_Mumbai_Sub_2022.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa/2021-22/DSA_2022_Mumbai_City.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa/2021-22/DSA_2022_Ratnagiri.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa/2021-22/DSA_Sindhudurg_2021-22.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa/2021-22/DSA_Palghar_2021-22.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa/2021-22/DSA_Nandurbar_2021-22.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa/2021-22/DSA_Dhule_2021-22.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa/2021-22/DSA_Jalgaon_2021-22.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa/2022-23/DSA_Nashik_2022-23.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa/2021-22/DSA_Ahmadnagar_2021-22.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa/2021-22/DSA_Pune_2021-22.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa/2021-22/DSA_Solapur_2021-22.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa/2021-22/DSA_Satara_2021-22.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa/2021-22/DSA_Kolhapur_2021-22.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa/2021-22/DSA_Sangli_2021-22.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa/2021-22/DSA_Nanded_2021-22.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa/2021-22/DSA_Hingoli_2021-22.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa/2021-22/DSA_Parbhani_2021-22.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Jalna_DSA_2022.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa/2021-22/DSA_Aurangabad_2021-22.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa/2021-22/DSA_Beed_2021-22.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa/2021-22/DSA_Latur_2021-22.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa/2021-22/DSA_Osmanabad_2021-22.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa/2021-22/DSA_Buldana_2021-22.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa/2022-23/DSA_Akola_2022-23.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa/2021-22/DSA_Washim_2021-22.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa/2021-22/DSA_Amravati_2021-22.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa/2021-22/DSA_Yavatmal_2021-22.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa/2021-22/DSA_Raigad_2021-22.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa/2021-22/DSA_Wardha_2021-22.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa/2021-22/DSA_2022_Bhandara.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa/2021-22/DSA_Gondiya_2021-22.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa/2021-22/DSA_Gadchiroli_2021-22.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa/2021-22/DSA_Nagpur_2021-22.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Chandrapur_DSA_2022.pdf",
"https://mahades.maharashtra.gov.in/files/publication/YAVATMAL_DSA%202021.pdf",
"https://mahades.maharashtra.gov.in/files/publication/DSA_2021_Thane.pdf",
"https://mahades.maharashtra.gov.in/files/publication/DSA_2021_Nandurbar.pdf",
"https://mahades.maharashtra.gov.in/files/publication/DSA_2021_Gadchiroli.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Buldhana_DSA_2021.pdf",
"https://mahades.maharashtra.gov.in/files/publication/DSA_2021_Mumbai_Suburban.pdf",
"https://mahades.maharashtra.gov.in/files/publication/DSA_2021_Ratnagir.pdf",
"https://mahades.maharashtra.gov.in/files/publication/DSA_2021_Ahemadnar.pdf",
"https://mahades.maharashtra.gov.in/files/publication/DSA_2021_Aurangabad.pdf",
"https://mahades.maharashtra.gov.in/files/publication/DSA_2021_Nanded.pdf",
"https://mahades.maharashtra.gov.in/files/publication/DSA_2021_Latur.pdf",
"https://mahades.maharashtra.gov.in/files/publication/DSA_2021_Bhandara.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa/2020-21/DSA_Ratnagiri_2020-21.pdf",
"https://mahades.maharashtra.gov.in/files/publication/DSA_Palghar_2021.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_%20Dhule%20_2021.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_raigad_2021.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Pune_2021.pdf",
"https://mahades.maharashtra.gov.in/files/publication/DSA_2021_Solapur.pdf",
"https://mahades.maharashtra.gov.in/files/publication/DSA_2021_Beed.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Amravati_2021.pdf",
"https://mahades.maharashtra.gov.in/files/publication/DSA_2021_nagpur.pdf",
"https://mahades.maharashtra.gov.in/files/publication/DSA_Thane_2020.pdf",
"https://mahades.maharashtra.gov.in/files/publication/DSA_2020_Mumbai_Suburban.pdf",
"https://mahades.maharashtra.gov.in/files/publication/DSA_Mumbai_City_2020.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_%20Ratnagiri%20_2019.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_%20Sindhudurg%20_2020.pdf",
"https://mahades.maharashtra.gov.in/files/publication/DSA_Palghar_2020.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Nandurbar_2020.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_%20Dhule%20_2020.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Jalgaon_2020.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_%20Nashik_2020.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_raigad_2020.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Ahmednagar_2020.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Pune_2020.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Solapur_2020.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Kolhapur_2020.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Hingoli_2020.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Parbhani_2020.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Jalna_2020.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Latur_2020.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Osmanabad_2020.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Buldhana_2020.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Akola_2020.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Washim_2020.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Amravati_2020.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Yavatmal_2020.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_%20Wardha%20_2020.pdf",
"https://mahades.maharashtra.gov.in/files/publication/DSA_2020_nagpur_revd.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Bhandara_2020.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Gondia_2020.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_%20Gadchiroli%20_2020.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Satara_2020.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Chandrapur_2020.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Thane_2019.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_MumbaiCity_2019.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_%20Ratnagiri%20_2019.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Sindhudurg_2019.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Nandurbar_2019.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_dhule_2019.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Jalgaon_2019.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Nashik_2019.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_raigad_2019.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Ahmednagar_2019.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Pune_2019.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Solapur_2019.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_%20Satara%20_2019.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Kolhapur_2019.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_%20Sangli%20_2019.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Nanded_2019.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Hingoli_2019.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Parbhani_2019.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_%20Jalna_rev%20_2019.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Aurangabad_2019.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Beed_2019.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Latur_2019.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_osmanabad_2019.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Buldhana_2019.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Akola_2019_rev.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Washim_2019.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Amravati_2019.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_yavatmal_2019.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_wardha_2019.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Nagpur_2019.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_bhandara%202019.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Gondia_2019_rev.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Gadchiroli_2019.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Chandrapur_2019.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Thane_2018.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Mumbaisub_2018.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Ratnagiri_2018.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_sindhudurg_2018.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_nandurbar_2018.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Dhule_2018.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Jalgaon_2018.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Nashik_2018_rev.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Raigad_2018.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Ahmednagar_2018.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Pune_2018.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Solapur_2018.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Satara_2018.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_kolhapur_2018.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Sangli_2018.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Nanded_2018_rev.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_hingoli_2018.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_parbhani_2018.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_MumbaiCity_2018.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Jalana_2018.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Aurangabad_2018.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Beed_2018_rev.pdf",
"https://mahades.maharashtra.gov.in/files/publication/DSA_LATUR_2018.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_osmanabad_2018.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Buldana_2018.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Akola_2018.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_washim_2018.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Amravati_2018.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_yavatmal_2018.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Wardha_2018.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Nagpur_2018.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Bhandara_2018.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Gondia_2018.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Gadchiroli_2018.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Chandrapur_2018.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Thane_2017.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_MumSub_2017.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_MumbaiCity_2017.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Ratnagiri_2017.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Sindhudurg_2017.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_%20Palghar%20_2017.pdf",
"https://mahades.maharashtra.gov.in/files/publication/DSA_Nandurbar_2017.pdf",
"https://mahades.maharashtra.gov.in/files/publication/DSA_Dhule_2017.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Jalagaon_2017.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Nashik_2017.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Ahmednagar_2017.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_raigad_2017.pdf.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Pune_2017.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Solapur_2017.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Satara_2017.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Kolhapur_2017.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Sangli_2017.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Nanded_2017.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Hingoli_2017.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Parbhani_2017.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Jalana_2017.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Aurangabad_2017.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Beed_2017.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Latur_2017.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Osmanabad_2017.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Buldhana_2017.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Akola_2017.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Washim_2017.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Amravati_2017.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Yavatmal_2017.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Wardha_2017.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Nagpur_2017.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Bhandara_2017.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Gondia_2017.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Gadchiroli_2017.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Chandrapur_2017.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Thane_2016.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_raigad_2016.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Ratnagiri_2016.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Sindhudurg_2016.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_palghar_2016.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Nandurbar_2016.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Dhule_2016.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Jalgaon_2016.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Nashik_2016.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Ahmednagar_2016.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_solapur_2016.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Satara_2016.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Kolhapur_2016.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Sangli_2016.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Nanded.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Hingoli_2016.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Parabhani_2016.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Jalna_2016.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Aurangabad_2016.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Beed_2016.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Latur_2016.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Osmanabad_2016.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Buldhana_2016.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Akola.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Washim_2016.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Amravati_2016.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Yavatmal_2016.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Wardha_2016.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Nagpur_2016.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Bhandara_2016.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Gondia_2016.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Gadachiroli_2016.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Chandrapur_2016.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_thane_2015.pdf",
"https://mahades.maharashtra.gov.in/files/publication/DSA_Mumbai Suburban_2015.pdf",
"https://mahades.maharashtra.gov.in/files/publication/DSA_Mumbaicity_2015.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_raigad_2015.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_ratnagiri_2015.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_sindhudurg_2015.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_nandurbar_2015.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_dhule_2015.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_jalgaon_2015.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_nashik_2015.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Ahmednagar_2015_Rev.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_satara_New2015.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Kolhapur_2015_Rev.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_sangli%20_2015.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_nanded%20_2015.PDF",
"https://mahades.maharashtra.gov.in/files/publication/dsa_hingoli%20_2015.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_parbhani%20_2015.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_jalana_2015.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_aurangabad_2015.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_beed%20_2015.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_latur%20_2015.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_osmanabad_2015.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_buldana_2015.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Akola_2015_Rev.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_washim%20_2015.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_amravati%20_2015.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_yavatmal_2015.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_wardha_2015.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_nagpur_2015.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_bhandara_2015.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_gondia_2015.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_gadchiroli_2015.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_chandrapur_2015.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_thane_2014.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Mumbai Suburban_2014.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_raigad_2014.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_ratnagiri_2014.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_sindhudurg_2014.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_nandurbar_2014.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_dhule_2014.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_jalgaon_2014.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_nashik_2014.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_ahmednagar_2014.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_pune_2014.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_solapur_2014.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_satara_2014.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_kolhapur_2014.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_sangli_2014.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_nanded_2014.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_hingoli_2014.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_parbhani_2014.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_jalna_2014.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_aurangabad_2014.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_beed_2014.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_latur_2014.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_osmanabad_2014.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_buldhana_2014.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_akola_2014.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_washim_2014.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_amravati_2014.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_yavatmal_2014.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_wardha_2014.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_nagpur_2014.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_bhandara_2014.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_gondia_2014.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_gadchiroli_2014.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_chandrapur_2014.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_thane_2013.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_raigad_2013.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_rartnagiri_2013.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_sindhudurg_2013.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_nandurbar_2013.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_dhule_2013.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_jalgaon_2013.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_nashik_2013.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_ahmednagar_2013.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_pune_2013.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_solapur_2013.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_satara_2013.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_kolhapur_2013.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_sangli_2013.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_nanded_2013.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_hingoli_2013.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_parbhani_2013.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_jalna_2013.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_aurangabad_2013.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_beed_2013.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_latur_2013.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_osmanabad_2013.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_buldhana_2013.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_akola_2013.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_washim_2013.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_amaravati_2013.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_yavatmal_2013.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_wardha_2013.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_nagpur_2013.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_bhandara_2013.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_gondia_2013.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_gadchiroli_2013.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_chandrapur_2013.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_thane_2012.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_raigad_2012.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_ratnagiri_2012.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_sindhudurg_2012.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_nandurbar_2012.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_dhule_2012.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_jalgaon_2012.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_nashik_2012.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_ahmednagar_2012.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_pune_2012.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_solapur_2012.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_satara_2012.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_kolhapur_2012.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_sangli_2012.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_nanded_2012.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_hingoli_2012.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_parbhani_2012.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_jalna_2012.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_aurangabad_2012.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_beed_2012.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_latur_2012.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_osmanabad_2012.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_buldhana_2012.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_akola_2012.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_washim_2012.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_amravati_2012.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_yavatmal_2012.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_wardha_2012.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_nagpur_2012.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_bhandara_2012.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_gondia_2012.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_gadchiroli_2012.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_chandrapur_2012.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_thane_2011.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_raigad_2011.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_ratnagiri_2011.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_sindhudurg_2011.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_nandurbar_2011.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_dhule_2011.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_jalgaon_2011.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_nashik_2011.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_ahmednagar_2011.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_pune_2011.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_solapur_2011.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_satara_2011.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_kolhapur_2011.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_sangli_2011.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_nanded_2011.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_hingoli_2011.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_parbhani_2011.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_jalna_2011.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_aurangabad_2011.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_beed_2011.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_latur_2011.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_osmanabad_2011.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_buldhana_2011.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_akola_2011.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_washim_2011.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_amravati_2011.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_yavatmal_2011.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_wardha_2011.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_nagpur_2011.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_Bhandara_2011.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_gondia_2011.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_gadchiroli_2011.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_chandrapur_2011.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_thane_2010.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_raigad_2010.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_ratnagiri_2010.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_sindhudurg_2010.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_nandurbar_2010.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_dhule_2010.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_jalgaon_2010.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_nashik_2010.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_ahmednagar_2010.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_pune_2010.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_solapur_2010.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_satara_2010.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_kolhapur_2010.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_sangli_2010.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_nanded_2010.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_hingoli_2010.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_parbhani_2010.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_jalna_2010.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_aurangabad_2010.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_beed_2010.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_latur_2010.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_osmanabad_2010.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_buldhana_2010.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_akola_2010.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_washim_2010.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_amravati_2010.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_yavatmal_2010.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_wardha_2010.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_nagpur_2010.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_bhandara_2010.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_gondia_2010.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_gadchiroli_2010.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_chandrapur_2010.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_thane_2009.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_raigad_2009.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_ratnagiri_2009.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_sindhudurg_2009.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_nandurbar_2009.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_dhule_2009.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_jalgaon_2009.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_nashik_2009.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_ahmednagar_2009.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_pune_2009.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_solapur_2009.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_satara_2009.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_kolhapur_2009.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_sangli_2009.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_nanded_2009.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_hingoli_2009.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_parbhani_2009.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_jalna_2009.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_aurangabad_2009.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_beed_2009.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_latur_2009.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_osmanabad_2009.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_buldhana_2009.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_akola_2009.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_washim_2009.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_amravati_2009.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_yavatmal_2009.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_wardha_2009.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_nagpur_2009.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_bhandara_2009.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_gondia_2009.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_gadchiroli_2009.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsa_chandrapur_2009.pdf",
"https://mahades.maharashtra.gov.in/files/publication/HINGOLI%20DSA%202021.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Parbhani_DSA_2021.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Sindhudurg_DSA%202021.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Mumbai_city_DSA_2021.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Jalgaon_DSA_2021.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Nashik_DSA_2021.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Kolhapur_DSA_2021.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Sangli_DSA_2021.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Akola_DSA_2021_01.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Gondia_DSA_2021_02.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Satara_DSA_2021_03.pdf",
"https://mahades.maharashtra.gov.in/files/publication/JALNA%20DSA%202021.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Osmanabad_DSA_2021.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Washim_DSA_2021.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Wardha_DSA_2021_04.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Chandrapur_DSA_2021.pdf",
"https://mahades.maharashtra.gov.in/files/publication/DSA-%20final%202020%20SANGLI.pdf",
"https://mahades.maharashtra.gov.in/files/publication/DSA%202019-20nanded.pdf",
"https://mahades.maharashtra.gov.in/files/publication/DSA%202020_Aurangabad.pdf",
"https://mahades.maharashtra.gov.in/files/publication/DSA%202020%20Final%20Beed-2.pdf",
"https://mahades.maharashtra.gov.in/files/publication/DSA%202018-19%20Mumbai%20suburban_compressed%20(1).pdf",
"https://mahades.maharashtra.gov.in/files/publication/DSA%20PALGHAR%202019%20FINAL.docx%20(1)%20(1)_compressed.pdf",
"https://mahades.maharashtra.gov.in/files/publication/DSA%20PALGHAR%202018.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Mumbai_suburban_2016R.PDF",
"https://mahades.maharashtra.gov.in/files/publication/Mumbai_City_2016.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Pune_DSA_2015_16.pdf",
"https://mahades.maharashtra.gov.in/files/publication/PALGHAR_DSA_%202015.pdf",
"https://mahades.maharashtra.gov.in/files/publication/PUNE%20DSA2015.pdf",
"https://mahades.maharashtra.gov.in/files/publication/SolapurDSA2015.pdf",
"https://mahades.maharashtra.gov.in/files/publication/DSA-2014-MUMBAI%20CITY.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Thanefinaldso_0607.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Raigad_2006-07.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Ratnagirifinaldso_0607.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Sindhudurgfinaldso_0607.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Nandurbar_2006-07.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Dhule_2006-07.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Nashik_2006-07.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Ahamadnagar_2006-07.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Pune_2006-07.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Satarafinaldso_0607.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Kolhapurfinaldso_0607.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Sangli_2006-07.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Nanded_2006-07.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Parbhanidsa_2006_07.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Jalna_2006-07.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Aurangabad_2006-07.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Beedfinaldso_0607.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Latur_2006-07.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Osmanabadfinaldso_0607.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Buldhana_2006-07.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Akola_2006-07.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Amravati_2006-07.pdf",
"https://mahades.maharashtra.gov.in/files/publication/yavatmaldsafinal_0607.pdf",
"https://mahades.maharashtra.gov.in/files/publication/wardhafinaldsa_0607.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Nagpur_2006-07.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Bhandaradsa_2006_07.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Gadchirolifinaldsa_0607.pdf",
"https://mahades.maharashtra.gov.in/files/publication/chandrapur_2006-07.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Nandurbarfinaldsa_0506.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Dhulefinaldso_0506.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Jalgaondsa_2005_06.pdf",
"https://mahades.maharashtra.gov.in/files/publication/dsanashik0506.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Buldhananfinaldso_0506.pdf",
"https://mahades.maharashtra.gov.in/files/publication/Akolafinaldso_0506.pdf",

    ]

    # Provide the folder path where you want to save the PDFs
    folder_path = r"C:\Users\Library\OneDrive\Documents\Yashveer\Directorate of Economics and Statistics\Maharashtra DES"

    download_pdfs_from_links(pdf_links, folder_path)
```
